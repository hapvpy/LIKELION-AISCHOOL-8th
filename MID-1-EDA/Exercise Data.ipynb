```python
# 등록취소, 신고취소, 허가취소 삭제

cancel_1 = df_gym['상세영업상태명'].isin(['등록취소', '신고취소', '허가취소'])
cancel_2 = df_golf['상세영업상태명'].isin(['등록취소', '신고취소', '허가취소'])
df_gym = df_gym[~cancel_1]
df_golf = df_golf[~cancel_2]
```

- 휴업자들은 주소지를 찾아보니 대부분 다른 사업명으로 다른 사업을 하고있었다.
- 그러므로 휴업은 폐업으로 이해하고, 휴업시작일자를 폐업일자로 설정하는 것으로 컨센서스

```python
# '휴업시작일자'에 있는 데이터를 '폐업일자'에 넣기.

df_gym['폐업일자'] = df_gym['휴업시작일자'].fillna(df_gym['폐업일자'])
df_golf['폐업일자'] = df_golf['휴업시작일자'].fillna(df_golf['폐업일자'])
```

```python
# '상세영업상태명' 통일하기
# '폐업'처리 = 폐업, 직권말소, 휴업, 영업장폐쇄 (2020년자료부터 남길 예정.)

df_gym['상세영업상태명'] = df_gym['상세영업상태명'].str.replace('직권말소|휴업|영업장폐쇄', '폐업')
df_golf['상세영업상태명'] = df_golf['상세영업상태명'].str.replace('직권말소|휴업|영업장폐쇄', '폐업')
```

```python
df_gym['상세영업상태명'].value_counts()
```

## 중부원점 위도/경도로 변환

```python
import pyproj

coord = list(zip(df_gym['좌표정보(y)'], df_gym['좌표정보(x)']))

def convert_coords(x, y):
    proj1 = pyproj.crs.CRS('epsg:2097')  # 중부원점 좌표계
    proj2 = pyproj.crs.CRS('epsg:4326')  # 위도·경도 좌표계
    transformer = pyproj.transformer.Transformer.from_crs(proj1, proj2)
    lon, lat = transformer.transform(x, y)
    return lon, lat

lon_lat = [convert_coords(x, y) for x, y in coord]

df_gym['위도'], df_gym['경도'] = zip(*lon_lat)
```

```python
import pyproj

coord = list(zip(df_golf['좌표정보(y)'], df_golf['좌표정보(x)']))

def convert_coords(x, y):
    proj1 = pyproj.crs.CRS('epsg:2097')  # 중부원점 좌표계
    proj2 = pyproj.crs.CRS('epsg:4326')  # 위도·경도 좌표계
    transformer = pyproj.transformer.Transformer.from_crs(proj1, proj2)
    lon, lat = transformer.transform(x, y)
    return lon, lat

lon_lat = [convert_coords(x, y) for x, y in coord]

df_golf['위도'],df_golf['경도'] = zip(*lon_lat)
```
