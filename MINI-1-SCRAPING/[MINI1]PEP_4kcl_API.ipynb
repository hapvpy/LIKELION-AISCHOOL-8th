```python
import folium
import pandas as pd
```

```python
# 2020년도 구청별 위경도 데이터프레임 만들기 
gps_2020 = {'강동구청' : ['37.53016268119579', '127.12379669318284'],
       '성동구청' : ['37.56344', '127.036958'],
       '광진구청' : ['37.53837', '127.082207'],
       '관악구청' : ['37.47836364595873', '126.951562139563'],
       '도봉구청' : ['37.66868465132199', '127.04721389543128'],
       '양천구청' : ['37.51697948762874', '126.86652122082332'],
       '서초구청' : ['37.48359038694852', '127.03273510990965'],
       '송파구청' : ['37.51458596941698', '127.10592053963026']}

df_gps_2020 = pd.DataFrame(gps_2020)
df_gps_2020 = df_gps_2020.T  # 데이터프레임 행,열 전치

df_gps_2020.columns = ["위도", "경도"]  # 위경도 컬럼 이름 지정
df_gps_2020.index.name = "구청"  # 구청 컬럼 지정
df_gps_2020.reset_index(inplace = True)  # 앞쪽에 새로운 인덱스 추가하기
df_gps_2020
```

```python
# 지도 시각화 준비
import requests
import json
```

```python
# 서울 행정구역 json raw파일(githubcontent)
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
# 2020년 1월 수치값 계산한 데이터 가져오기
df_2020_data2 = df_2020_data.loc[df_2020_data['수록시점'] == '202001', :].copy()
df_2020_data2
```

```python
df_2020_data3 = df_2020_data2[['분류값명2', '수치값']].copy()
df_2020_data3 = df_2020_data3.reset_index(drop = True)
df_2020_data3
```

```python
df_gps_data = df_gps_2020.join(df_2020_data3, how = 'left')
df_gps_data
```

```python
df_gps_data.info()
```

```python
df_gps_data = df_gps_data.astype({'수치값':'int64'})
```

```python
import plotly.express as px
from folium.plugins import MarkerCluster

# 처음 지도 위치 잡기 
m_2020 = folium.Map(
    location=[37.559819, 126.963895],  # 서울 좌표 
    zoom_start=11, 
)

# 행정단위 별 시각화
folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(m_2020)


# 구청별 위경도를 이용한 마크 표시 
marker_cluster = MarkerCluster().add_to(m_2020)

for g, v, lat, long in zip(df_gps_data["분류값명2"], df_gps_data['수치값'], df_gps_data['위도'], df_gps_data['경도']):
    folium.Marker([lat, long],
                  popuo = g,
                  tooltip = g + ' : ' + str(v),
                  icon = folium.Icon(color="green")).add_to(marker_cluster)

# 지도 채도 변화
m_2020.choropleth(geo_data = seoul_geo,
                  data = df_gps_data,
                  columns = ('분류값명2', '수치값'),
                  fill_color='OrRd', # 색상 변경도 가능하다
                  fill_opacity=0.8, # 투명도
                  line_opacity=1,
                  key_on= 'feature.properties.name',
                  legend_name="2020년 1월 구별 빈집 수"
                )

m_2020
```
